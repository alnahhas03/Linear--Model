```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = torch.randn(100, 1)*10
y = x + 3*torch.randn(100, 1)
plt.plot(x.numpy(), y.numpy(), 'o')
plt.ylabel('y')
plt.xlabel('x')
```

```python
class LR(nn.Module):
  def __init__(self, input_size, output_size):
    super().__init__()
    self.linear = nn.Linear(input_size, output_size)
  def forward(self, x):
    pred = self.linear(x)
    return pred
```

```python
torch.manual_seed(1)
model = LR(1, 1)
```

```python
[w, b] = model.parameters()
def get_params():
  return (w.item(), b.item())
```

```python
def plot_fit(title):
  plt.title = title
  w1, b1 = get_params()
  x1 = np.array([-30, 30])
  y1 = w1*x1 + b1
  plt.plot(x1, y1, 'r')
  plt.scatter(x, y)
  plt.show()
```

```python
plot_fit("fit_itit")
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr = 0.01)
```

```python
epochs = 100
losses = []
for i in range(epochs):
   y_pred = model.forward(x)
   loss = criterion(y_pred, y)
   print("epoch:", i, "loss:", loss.item())
   losses.append(loss)
   optimizer.zero_grad()
   loss.backward()
   optimizer.step()
```

```python
plt.plot(range(epochs), [loss.detach().numpy() for loss in losses])
plt.ylabel('Loss')
plt.xlabel('epoch')
```
